# Data Visualization
- Line Plot

```python
# import libraraies
import seaborn as sns
import matplotlib.pyplot as plt

# Load Data Set
flower= sns.load_dataset("iris")
flower

# Draw a Line Plot
sns.lineplot(x="petal_length",y="petal_width",data=flower)
plt.show()
```

## Adding Title

```python
# import libraraies
import seaborn as sns
import matplotlib.pyplot as plt

# Load Data Set
flower= sns.load_dataset("iris")
flower

# Draw a Line Plot
sns.lineplot(x="petal_length",y="petal_width",data=flower)
plt.title("Plot of Flower")
plt.show()
```

### Adding Limits

```python
# import libraraies
import seaborn as sns
import matplotlib.pyplot as plt

# Load Data Set
flower= sns.load_dataset("iris")
flower

# Draw a Line Plot
sns.lineplot(x="petal_length",y="petal_width",data=flower)
plt.title("Plot of Flower")
plt.xlim(0)
plt.ylim(-1)
plt.show()
```

## Size of Figure

```python
# import libraraies
import seaborn as sns
import matplotlib.pyplot as plt

# Load Data Set
flower= sns.load_dataset("iris")
flower

# change fig size
plt.figure(figsize=(10,8))

# Draw a Line Plot
sns.lineplot(x="petal_length",y="petal_width",data=flower)
plt.title("Plot of Flower")
plt.show()
```
